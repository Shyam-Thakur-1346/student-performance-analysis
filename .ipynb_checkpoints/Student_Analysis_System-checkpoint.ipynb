```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("student_marks.csv")
print(df)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df[["maths","physics","chemistry"]] = df[["maths","physics","chemistry"]].fillna(
    df[["maths","physics","chemistry"]].mean()
)
```

```python
df
```

```python
def total_marks(row):
    return row["maths"]+row["physics"]+row["chemistry"]
```

```python
df["total_marks"]=df.apply(total_marks,axis=1)
```

```python
df
```

```python
df=df.round({
    "maths":2,
    "physics":2,
    "chemistry":2,
    "total_marks":2
})
```

```python
df
```

```python
def percentage(row):
    return row["total_marks"]/300 *100
```

```python
df["percentage"]=df.apply(percentage,axis=1)
```

```python
df
```

```python
df=df.round({"percentage":2})
```

```python
df
```

FIND TOPPER ?

```python
def get_topper(df):
    return df.loc[df["percentage"].idxmax()]
topper=get_topper(df)
print(f"Topper Name : {topper["name"]},\nPercentage :{topper["percentage"]}")

```

FAILED STUDENT ?


```python
failed_students=df[
    (df["maths"]<50) |
    (df["chemistry"]<50) |
    (df["physics"]<50)
]
print(failed_students)
```

AVERAGE MARKS PER SUBJECT

```python
average_marks=df[["maths","physics","chemistry"]].mean()
print(average_marks.round(2))
```

NUMPY: CONVERT MARKS INTO NUMPY ARRAYS

```python
marks_array = df[["maths","physics","chemistry"]].to_numpy()
print(marks_array)
```

```python
np.mean(marks_array,axis=0)
```

```python
np.max(marks_array,axis=0)
```

```python
np.min(marks_array,axis=0)
```

```python
np.std(marks_array,axis=0)      # standard deviation
```
